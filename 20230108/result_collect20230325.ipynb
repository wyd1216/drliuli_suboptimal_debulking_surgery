```python
import pandas as pd
```

```python
# clinical results
train_df = pd.read_csv('./data1/train_results.csv')
train_df.rename(columns={'train_pid':'pid', 'train_pred':'y_score1', 'train_label':'y_true'}, inplace=True)
train_df['y_score0'] = train_df['y_score1'].apply(lambda x: 1-x)
train_df = train_df.iloc[:,1:]
train_df['dataset']='training'
train_df.head()
```

```python
# clinical results
test_df = pd.read_csv('./data1/test_results.csv')
test_df.rename(columns={'test_pid':'pid', 'test_pred':'y_score1', 'test_label':'y_true'}, inplace=True)
test_df['y_score0'] = test_df['y_score1'].apply(lambda x: 1-x)
test_df = test_df.iloc[:,1:]
test_df['dataset']='test'
test_df.head()
```

```python
# clinical results
external_df = pd.read_csv('./data1/extest_results.csv')
extest_df.rename(columns={'extest_pid':'pid', 'extest_pred':'y_score1', 'extest_label':'y_true'}, inplace=True)
extest_df['y_score0'] = extest_df['y_score1'].apply(lambda x: 1-x)
extest_df = extest_df.iloc[:,1:]
extest_df['dataset']='external'
extest_df.head()
```

```python
cli_all_df = pd.concat([train_df, test_df, extest_df], axis=0)
threshold = 0.51685315972792
cli_all_df.insert(loc=2, column='y_pred', value=cli_all_df['y_score1'].apply(lambda x: 0 if x<threshold else 1))
cli_all_df.head()
```

```python
cli_all_df.to_csv('./data_20230325/clinical_origin.csv', index=0)
```
